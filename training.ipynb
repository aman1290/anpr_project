```python
import os
import urllib.request

# Create a directory for the weights in the current working directory
weights_dir = os.path.join(os.getcwd(), "weights")
os.makedirs(weights_dir, exist_ok=True)

# URLs of the weight files
urls = [
    "https://github.com/sunsmarterjie/yolov12/releases/download/turbo/yolov12n.pt",
    "https://github.com/sunsmarterjie/yolov12/releases/download/turbo/yolov12s.pt",
    "https://github.com/sunsmarterjie/yolov12/releases/download/turbo/yolov12m.pt",
    "https://github.com/sunsmarterjie/yolov12/releases/download/turbo/yolov12l.pt",
    "https://github.com/sunsmarterjie/yolov12/releases/download/turbo/yolov12x.pt",
]

# Download each file
for url in urls:
    file_name = os.path.join(weights_dir, os.path.basename(url))
    urllib.request.urlretrieve(url, file_name)
    print(f"Downloaded {file_name}")
```
